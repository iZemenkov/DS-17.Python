# Сбор данных из интернета

В библиотеке pandas есть функция `read_html()`, которая делает запрос к странице, находит в HTML-разметке теги `<table>`, преобразовывает эти данные в датафреймы и возвращает список таблиц.

```python
import pandas as pd
df = pd.read_html('https://ru.wikipedia.org/wiki/%D0%A6%D0%95%D0%A0%D0%9D')
```

```python
len(df)
```

```python
df[0] # подставьте индексы от 0 до 11, увидите все таблицы, которые были на странице
```

Если мы знаем, что ищем, среди всех таблиц на странице, то можем использовать аргумент `match`. Обратите внимание, что даже если на странице всего одна таблица, функция `read_html` вернет ее как список из одного элемента, поэтому в конце обязательно использовать индекс `[0]`.

```python
pd.read_html('https://ru.wikipedia.org/wiki/%D0%A6%D0%95%D0%A0%D0%9D', match='пожертвование')[0]
```

ПОДСКАЗКА

2. Удалите все строки, в которых еще не указан счет (игры пройдут позже)

```python
df.reset_index(drop=True, inplace=True)
df.dropna(axis=1, inplace=True)
df.dropna(axis=0, inplace=True)
df
```

В полученном датафрейме нужно вытащить все значения, разделить команды и голы по разделителям, сохранить их как два списка. Далее на основе полученных данных сделать новый датафрейм с этими столбцами.

```python
goals = df['Счет'].apply(lambda x: x.split(':')).sum()
goals = list(map(int, goals))
print(goals)
```

```python
teams = df['Матч'].apply(lambda x: x.split('\xa0– ')).sum()
print(teams)
```

Сделайте группирующую таблицу по количеству голов для каждой команды.

```python
df2 = pd.DataFrame(teams, goals)
```

```python
# df2.reset_index(inplace=True)
# df2.columns = ['goal','team']
# df2.drop('l',inplace=True, axis=1)
df2.groupby('team')[['goal']].sum().sort_values('goal', ascending=False)
```

### Задача со звездочкой

Определите, в какие дни недели чаще всего проходят матчи. Вам понадобится тип данных [datetime](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html).

```python
df['Дата, время'].apply(lambda x: x[0:5] + '.2024' if ',' in x else x)
```

```python
df['date'] = df['Дата, время'].apply(lambda x: x[0:5] + '.2024' if ',' in x else x)
```

```python
df
```

```python
df.date = pd.to_datetime(df.date, dayfirst=True)
```

```python
df['dayname'] = df.date.dt.day_name()
```

```python
df.groupby('dayname')[['Матч']].count().sort_values('Матч', ascending=False)
```

```python
df['dayname'].value_counts(normalize=True) * 100
```
